```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs # Organiza HTML´s
```

```python
html="<!DOCTYPE html><html><head><title>Page Title</title></head><body><h3><b id='boldest'>Lebron James</b></h3><p> Salary: $ 92,000,000 </p><h3> Stephen Curry</h3><p> Salary: $85,000, 000 </p><h3> Kevin Durant </h3><p> Salary: $73,200, 000</p></body></html>"

soup = bs(html, 'html5lib') # Para criar estrutura em arvore do html
```

```python
print(soup.prettify()) # Apresenta HTML em estrutura aninhada
```

```python
tag_object=soup.title # tittle é tag de html
print("tag object:",tag_object) 
```
